# Open Petri Nets

```julia
using Catlab, Catlab.CategoricalAlgebra
using AlgebraicPetri
```

### Infection Process
First we define a (closed) Petri net representing an infection process. 

```julia
inf_net = LabelledPetriNet([:S, :I],
    :β => ((:S, :I) => (:I, :I))
)

Graph(inf_net)
```

We turn this Petri net into an *open* Peti net by specifying two legs and the species that they expose. 

```julia
open_inf = Open([:S, :I], inf_net, [:I]);
```

We can visualize the two legs as follows.

```julia
Graph(legs(open_inf)[1])
```

```julia
Graph(legs(open_inf)[2])
```

### Recovery Process
Let's repeat these steps for a recovery process.

```julia
rec_net = LabelledPetriNet([:I, :R],
    :γ => (:I => :R)
)

open_rec = Open([:I], rec_net, [:I, :R])

Graph(rec_net)
```

```julia
Graph(legs(open_rec)[1])
```

```julia
Graph(legs(open_rec)[2])
```

# Composing Open Systems

Now we want to compose our two open systems `open_inf_net` and `open_rec_net`.

```julia
open_sir = compose(open_inf, open_rec);
Graph(open_sir)
```

```julia
Graph(legs(open_sir)[1])
```

```julia
Graph(legs(open_sir)[2])
```
